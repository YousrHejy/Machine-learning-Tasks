### Question 10
1. Load Iris dataset from sklearn.

```python
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.concat([pd.DataFrame(iris.data), pd.DataFrame(iris.target)], axis=1)
df.columns = np.append(iris.feature_names, 'target')
df.columns
```

```python
X = df.drop('target', axis =1)
y = df['target']
```

```python
x_train, x_test, y_train, y_test =  train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from matplotlib import cm

cmap = cm.get_cmap('gnuplot')
scatter = pd.plotting.scatter_matrix(x_train, c= y_train, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(10,10), cmap=cmap)
```

```python
x_train = x_train.drop(['sepal length (cm)', 'petal width (cm)'], axis =1)
x_test = x_test.drop(['sepal length (cm)', 'petal width (cm)'], axis =1)
x_train
```

```python
k_range = range(1,20)
scores = []
scores_train = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    scores_train.append(knn.score(x_train, y_train))
    scores.append(knn.score(x_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20]);
print(scores_train)
print('##########################################')
print(scores)
```

```python
knn_clf = KNeighborsClassifier(n_neighbors = 9)
knn_clf.fit(x_train, y_train)
```

```python
y_pred = knn_clf.predict(x_test)
knn_clf.score(x_train, y_train), knn_clf.score(x_test, y_test)
```
